```python
import numpy as np
import pandas as pd
from datetime import datetime

df = pd.read_csv("../csv/dht-temp.csv");
print(df.shape)
df.head(3)
```

```python
df.drop(['item_data_description','item_image'], axis=1,inplace=True)
# df.drop('item_image', axis=1, inplace=True)
df.head()
```

```python
df.drop(['item_id','client_id'], axis=1,inplace=True)
df.head()
```

```python
print(df[df['item_name']=='dht/tc'].dropna())
```

```python
df['item_time'] = pd.to_datetime(df['item_time'], format='%Y-%m-%d %H:%M:%S')
df['item_time']=df['item_time'].apply(lambda x: x.replace(tzinfo=None))
df['item_time']=df['item_time'].values.astype('datetime64[s]')
print(df['item_time'])
```

```python


dfNew = df[df['item_name']=='dht/tc'].dropna()
df.to_csv('dht-temp.csv')
```

```python
dfNewHum = df[df['item_name']=='dht/hum'].dropna()
dfNewHum.to_csv('dht-hum.csv')
```

```python
dfNewGas = df[df['item_name']=='gas'].dropna()
dfNewGas.to_csv('gas.csv')
```

```python
dfNewLDR = df[df['item_name']=='ldr'].dropna()
dfNewLDR.to_csv('ldr.csv')
```

```python
dfNewPIR = df[df['item_name']=='pir'].dropna()
dfNewPIR.to_csv('pir.csv')
```
